```python
# --- 1. Import Library ---
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from minisom import MiniSom
import matplotlib.pyplot as plt
```

```python
# --- 2. Load Dataset ---
df = pd.read_csv("data_pkk.csv")
```

```python
# --- 3. Pilih kolom Likert untuk SOM ---
columns_for_som = [
    ' Saya percaya belanja online di Shopee/Tokopedia lebih hemat biaya.',
    'Saya menganggap penting kenyamanan dalam proses belanja online.',
    'Saya lebih suka belanja online karena banyak diskon dan promo.',
    'Saya lebih memilih platform online yang sudah banyak dipercaya orang.',
    ' Saya merasa lebih aman belanja langsung di toko fisik.',
    'Saya lebih puas melihat dan memegang langsung produk sebelum membeli.',
    'Saya tetap menganggap belanja offline penting meskipun belanja online mudah.',
    'Saya menganggap toko fisik lebih jujur daripada toko online.'
]
# Bersihkan spasi dari nama kolom
df.columns = [col.strip() for col in df.columns]
columns_for_som = [col.strip() for col in columns_for_som]

# Ganti '-' dengan NaN dan ubah ke float
df_cleaned = df[columns_for_som].replace('-', np.nan).apply(pd.to_numeric, errors='coerce')
df_cleaned = df_cleaned.dropna()
```

```python
# Normalisasi
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df_cleaned)
```

```python
# --- 5. Inisialisasi dan latih SOM ---
n_samples = data_scaled.shape[0]
n_neurons = int(np.ceil(5 * np.sqrt(n_samples)))  # total neuron
side = int(np.ceil(np.sqrt(n_neurons)))           # ukuran grid mendekati persegi

som = MiniSom(x=side, y=side, input_len=data_scaled.shape[1], sigma=1.0, learning_rate=0.5)
som.random_weights_init(data_scaled)
som.train_random(data_scaled, 1000)
```

```python
# --- 6. Visualisasi Posisi Responden Menggunakan U-Matrix ---
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.title("Pemetaan Responden di Grid SOM", fontsize=16)

# Background: U-Matrix
plt.pcolor(som.distance_map().T, cmap='coolwarm', edgecolors='k', linewidths=0.5)
plt.colorbar(label='Jarak antar Neuron')

# Anotasi setiap responden
for i, x in enumerate(data_scaled):
    w = som.winner(x)
    plt.text(w[0] + 0.5, w[1] + 0.5, str(i),
             ha='center', va='center',
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'),
             fontsize=9, color='black')

plt.xticks(np.arange(0, som.get_weights().shape[0] + 1))
plt.yticks(np.arange(0, som.get_weights().shape[1] + 1))
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.gca().invert_yaxis()  # agar (0,0) di kiri atas
plt.tight_layout()
plt.show()
```

```python
# --- 7. Hitung Distribusi Responden per Neuron ---
from collections import Counter

# Ambil koordinat neuron (BMU) untuk tiap responden
bmu_coords = [som.winner(x) for x in data_scaled]

# Hitung jumlah responden per neuron
bmu_counts = Counter(bmu_coords)

# Visualisasi distribusi jumlah responden di grid
plt.figure(figsize=(8, 8))
for coord, count in bmu_counts.items():
    plt.text(coord[0] + 0.5, coord[1] + 0.5, str(count),
             ha='center', va='center',
             bbox=dict(facecolor='lightblue', edgecolor='black'))
plt.pcolor(som.distance_map().T, cmap='coolwarm')
plt.colorbar()
plt.title("Jumlah Responden per Neuron")
plt.grid()
plt.show()
```

```python
# --- 8. Profilisasi Neuron: Karakteristik Rata-Rata per Klaster
from collections import Counter

# 1. Hitung neuron pemenang (BMU) untuk tiap responden
bmu_coords = [som.winner(x) for x in data_scaled]
bmu_counts = Counter(bmu_coords)

# 2. Hubungkan neuron dengan indeks responden yang masuk
neuron_to_indices = {}
for i, coord in enumerate(bmu_coords):
    neuron_to_indices.setdefault(coord, []).append(i)

# 3. Ambil 5 neuron terpadat
most_common_neurons = bmu_counts.most_common(10)

# 4. Cetak profil rata-rata untuk tiap neuron
for coord, count in most_common_neurons:
    print(f"\n📍 Neuron {coord} ({count} responden):")
    indices = neuron_to_indices[coord]
    profil = df_cleaned.iloc[indices].mean().round(2)
    print(profil)

```

Silhouette Score mengukur seberapa mirip data dalam kluster dibandingkan dengan kluster lain. Semakin tinggi (maksimum 1), semakin baik.

Davies-Bouldin Index mengukur seberapa baik kluster dipisahkan. Semakin kecil, semakin baik.

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
import numpy as np

# Ambil koordinat neuron (BMU) sebagai label cluster
cluster_labels = [som.winner(x) for x in data_scaled]

# Konversi koordinat neuron ke label integer
unique_coords = list(set(cluster_labels))
coord_to_label = {coord: i for i, coord in enumerate(unique_coords)}
numeric_labels = [coord_to_label[coord] for coord in cluster_labels]

# Hitung metrik evaluasi clustering
sil_score = silhouette_score(data_scaled, numeric_labels)
db_index = davies_bouldin_score(data_scaled, numeric_labels)

print(f"Silhouette Score: {sil_score:.4f}")
print(f"Davies-Bouldin Index: {db_index:.4f}")
```
